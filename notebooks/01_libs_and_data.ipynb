### Select-Apply-Combine

_... kinda of like "map-reduce"_

```python
data['gdpPercap_1952'] > data['gdpPercap_1952'].mean()
```

```python
data.apply(lambda x: x > x.mean())
```

```python
mask = data['gdpPercap_1952'] > data['gdpPercap_1952'].mean()
data.loc[mask, 'gdpPercap_1952']
```

```python
mask = data.apply(lambda x: x > x.mean())
score = mask.aggregate('sum', axis=1) / len(data.columns)
score
```
